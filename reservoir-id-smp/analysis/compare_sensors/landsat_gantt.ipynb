```python
import pandas as pd
from matplotlib import patches
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.dates as mdates


color_list = [ '#f9dc5c','#fda0bb', '#8da0cb', '#fc8d62','#66c2a5']
#l
# Example data
#

#Original data
df = pd.DataFrame(
    {'Task': ['Sentinel', 'LS9','Landsat 8','Landsat 7', 'Landsat 5'],
    #  'Start': ['1984-03-01', '1999-01-01', '2013-04-01', '2017-03-01'],
    #         'End': ['2012-05-01', '2022-04-01', '2025-01-01', '2025-01-01'],
     'Start': [2015, 2022, 2013, 1999, 1984],
            'End': [2025, 2025, 2025, 2022, 2012]
     }
)

height = 0.9

f, ax = plt.subplots(figsize=(12, 1.7))
ax.xaxis.grid(True)
for idx in range(len(df)):
    y0 = (idx + 1) - height / 2
    x0 = df.iloc[idx].Start
    width = df.iloc[idx].End - x0
    # if not width:
    #      width = pd.Timedelta(days=0.1)
    if idx==0:
        ax.add_patch( patches.Rectangle((x0, y0), width, height, color=color_list[idx], zorder=4, hatch='/'))
    else:
        ax.add_patch( patches.Rectangle((x0, y0), width, height, color=color_list[idx], zorder=4))
    ax.hlines(y0 + height / 2,
              xmin=df.iloc[idx].Start,
              xmax=x0,
              color='black', linestyles='-', linewidth=0.5)
    middle = np.mean([x0, df.iloc[idx].End])
    ax.text(middle,y0+0.25, df.iloc[idx].Task, ha='center', zorder=5)


ax.set_xlabel('Year')
ax.set_xticks(np.arange(1985, 2026, 5))
ax.set_xlim(1982, 2025)
ax.set_ylabel('Satellite')
ax.set_yticks([])
plt.show()
```
